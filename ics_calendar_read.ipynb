```python
import requests

def get_user_calendar_events(user_id, calendar_id, access_token, from_dt, until_dt):
    """
    Works API 의 GET /users/{userId}/calendars/{calendarId}/events 를 호출하여
    일정 목록을 조회합니다.
    
    :param user_id: 대상 사용자 ID (예: "userf7da-f82c-4284-13e7-030f3b4c756x")
    :param calendar_id: 대상 캘린더 ID (예: "calendar-96b8-4c74-8277-7953e0b4604a")
    :param access_token: OAuth2.0 Access Token (Bearer)
    :param from_dt: 조회 범위의 시작 시간 (ISO8601 포맷 예: "2025-02-01T00:00:00Z")
    :param until_dt: 조회 범위의 종료 시간 (ISO8601 포맷 예: "2025-02-28T23:59:59Z")
    :return: API 응답(JSON)
    """
    
    # 요청 URL 생성
    url = f"https://www.worksapis.com/v1.0/users/{user_id}/calendars/{calendar_id}/events"
    
    # 헤더에 인증 토큰 설정 (Bearer {access_token})
    headers = {
        "Authorization": f"Bearer {access_token}",
    }
    
    # Query Parameters
    # (fromDateTime~untilDateTime 기간이 최대 31일로 제한된다고 명시되어 있으므로 주의)
    params = {
        "fromDateTime": from_dt,
        "untilDateTime": until_dt,
    }
    
    # GET 요청
    response = requests.get(url, headers=headers, params=params)
    
    # 요청 실패 시 예외 발생
    response.raise_for_status()
    
    # JSON 데이터 반환
    return response.json()

user_id = "8zESjIVD54_8K5PA_dIu"
calendar_id = "8zESjIVD54_8K5PA_dIu"
access_token = "wyWAH1uLEY"  # 실제 액세스 토큰으로 교체

# ISO8601 형식 예시 (2025년 2월 1일 00:00 ~ 2월 28일 23:59:59)
from_datetime = "2025-01-01T00:00:00Z"
until_datetime = "2025-01-31T23:59:59Z"

# API 호출
events_data = get_user_calendar_events(user_id, calendar_id, access_token, from_datetime, until_datetime)
```

```python
import requests

def get_personal_calendars(user_id, access_token, count=50, cursor=None):
    """
    해당 사용자의 개인 캘린더 목록을 가져옵니다.
    :param user_id: 대상 사용자 ID
    :param access_token: OAuth 2.0 Access Token
    :param count: 조회 개수 (기본 50, 최대 50)
    :param cursor: 페이지네이션을 위한 커서 (없으면 None)
    :return: 응답 JSON(dict)
    """
    base_url = "https://www.worksapis.com/v1.0"
    endpoint = f"/users/{user_id}/calendar-personals"
    
    headers = {
        "Authorization": f"Bearer {access_token}",
    }
    
    params = {
        "count": count,
    }
    if cursor:
        params["cursor"] = cursor
    
    response = requests.get(base_url + endpoint, headers=headers, params=params)
    response.raise_for_status()  # 4xx / 5xx 에러 시 예외 발생
    return response.json()


# 예시 값
user_id = "abcd00pi@seowoohr.com"
access_token = "kr1NW1Tck52WHRwdUU5azlGaw=="  # 실제 발급받은 토큰으로 교체

# 캘린더 목록 조회
result_json = get_personal_calendars(user_id, access_token, count=50)

# calendars 배열에서 calendarId 값 추출
calendars = result_json.get("calendars", [])
for cal in calendars:
    cal_id = cal.get("calendarId")
    cal_summary = cal.get("summary")
    print(f"Calendar ID: {cal_id}, summary: {cal_summary}")

# hasNext와 cursor를 이용해 페이징 처리 가능
has_next = result_json.get("hasNext", False)
next_cursor = result_json.get("cursor")
if has_next and next_cursor:
    print(f"다음 페이지 존재. cursor={next_cursor}")
    # 필요 시 추가 호출
    # next_result = get_personal_calendars(user_id, access_token, count=50, cursor=next_cursor)
    # ...
```
