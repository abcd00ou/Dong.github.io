```python
-- SQLite SQL Schema for Seowoo HR System

-- 1. 현장 메타 정보
CREATE TABLE worksite_meta (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    site_name TEXT,
    photo_folder_path TEXT,
    meal_price INTEGER,
    bank_name TEXT,
    account_holder TEXT,
    account_number TEXT
);

-- 2. 작업 단가 테이블
CREATE TABLE unit_price (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    worksite_id INTEGER,
    floor TEXT,
    section TEXT,
    task_name TEXT,
    contract_type TEXT,
    category TEXT,
    unit TEXT,
    unit_price INTEGER,
    FOREIGN KEY (worksite_id) REFERENCES worksite_meta(id)
);

-- 3. 근로자 정보
CREATE TABLE worker (
    worker_id TEXT PRIMARY KEY,
    name TEXT,
    english_name TEXT,
    nationality TEXT,
    visa_type TEXT,
    birth_number TEXT,
    gender TEXT,
    age INTEGER,
    passport_number TEXT,
    bank_name TEXT,
    account_number TEXT,
    phone_number TEXT,
    depositor_name TEXT,
    team_key TEXT,
    payment_type TEXT
);

-- 4. 일별 공수 기록
CREATE TABLE labor_record (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date TEXT,
    worker_id TEXT,
    work_type TEXT,
    labor_unit REAL,
    real_output REAL,
    is_assistant TEXT,
    meal_outside REAL,
    meal_inside REAL,
    remarks TEXT,
    FOREIGN KEY (worker_id) REFERENCES worker(worker_id)
);

-- 5. 일별 작업 인력 상세
CREATE TABLE daily_workforce (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date TEXT,
    worksite_id INTEGER,
    task_key TEXT,
    floor TEXT,
    section TEXT,
    task_name TEXT,
    category TEXT,
    contract_type TEXT,
    total_quantity REAL,
    unit TEXT,
    accumulated_progress REAL,
    remaining_quantity REAL,
    unit_price INTEGER,
    task_description TEXT,
    worker_name TEXT,
    individual_labor REAL,
    contribution_ratio REAL,
    adjusted_contribution_ratio REAL,
    progress_quantity_per_person REAL,
    wage_per_person REAL,
    note TEXT,
    photo1 TEXT,
    photo2 TEXT,
    pouring_flag TEXT
);

-- 6. 월별 작업 기록
CREATE TABLE monthly_work_log (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    floor TEXT,
    section TEXT,
    task_name TEXT,
    quantity REAL,
    unit TEXT,
    unit_price INTEGER,
    contract_amount INTEGER,
    total_progress_amount INTEGER,
    progress_rate REAL
    -- 추후 필요 시 기타 열 추가 가능
);

-- 7. 한글-영어 컬럼명 매핑 테이블
CREATE TABLE column_translation (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    table_name TEXT,
    korean_name TEXT,
    english_name TEXT,
    description TEXT
);
```
